```
import pandas as pd 
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
%matplotlib inline
```

```
dataset['Attrition'] = dataset['Attrition'].map({'Yes' : 1, 'No' : 0})
```

```
dataset.T
```

# **1.  Correlation between Attrition and Age**

```
stats, p = pearsonr(dataset.Attrition, dataset.Age)
print(stats, p)
```

# **Hypothesis between Attrition and Age**
null hypothesis (H0) = Age and attrition have no correlation .

alternate hypothesis (H1) = Age and attrition have a correlation between them .

r = -0.15920500686577949

p = 1.996801615887198e-26

r is negative so there is a negative relation between Age and attrition


and as p <0.05 so null hypothesis (H0) is rejected and we find that there is a weak negative correlation between attrition and age

```
plt.scatter(dataset['Attrition'], dataset['Age'])
```
