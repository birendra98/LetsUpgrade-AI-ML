# Random Forest-

Importing packages

```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn import tree
```

```python
from sklearn import preprocessing
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data_train=pd.read_csv("general_data.csv")
```

```python
data_train
```

```python
data_train.describe().T
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
data_train["Attrition"]=label_encoder.fit_transform(data_train["Attrition"])
```

```python
data_train["BusinessTravel"]=label_encoder.fit_transform(data_train["BusinessTravel"])
```

```python
data_train["EducationField"]=label_encoder.fit_transform(data_train["EducationField"])
```

```python
data_train["Gender"]=label_encoder.fit_transform(data_train["Gender"])
```

```python
data_train["Over18"]=label_encoder.fit_transform(data_train["Over18"])
```

```python
data_train["Department"]=label_encoder.fit_transform(data_train["Department"])
```

```python
data_train["JobRole"]=label_encoder.fit_transform(data_train["JobRole"])
data_train["MaritalStatus"]=label_encoder.fit_transform(data_train["MaritalStatus"])
```

```python
data_train
```

```python
data_train.describe().T
```

```python
data_train.isna().sum()
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
features=["Age","BusinessTravel","Department","DistanceFromHome","Education","EducationField","EmployeeCount","Gender","JobLevel","JobRole","MaritalStatus","MonthlyIncome","YearsSinceLastPromotion","PercentSalaryHike","YearsAtCompany"]
```

```python
rf_model.fit(X=data_train[features],y=data_train["Attrition"])
```

```python
print("OOB accuracy")
print(rf_model.oob_score_)
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

Age,MonthlyIncome,YearsatCompany these three are the most important IDV

# Decision Tree

```python
from sklearn import tree
```

```python
tree_model=tree.DecisionTreeClassifier()
```

```python
tree_model.fit(X = pd.DataFrame(data_train["Age"]) , y =pd.DataFrame(data_train["Attrition"]))
```

```python
with open("AtrrDtree1.dot" , 'w') as f:
    f = tree.export_graphviz(tree_model , feature_names = ["Age"] , out_file =f)
```

```python
tree_model.fit(X = pd.DataFrame(data_train["MonthlyIncome"]), y =pd.DataFrame(data_train["Attrition"]))
```

```python
with open("AtrrDtree2.dot" , 'w') as f:
    f = tree.export_graphviz(tree_model , feature_names = ["MonthlyIncome"] , out_file =f)
```
