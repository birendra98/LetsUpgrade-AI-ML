# Attrition logistic regression

```python
import pandas as pd
```

```python
from sklearn import preprocessing
```

```python
data_train=pd.read_csv("general_data.csv")
```

```python
data_train
```

```python
data_train.describe().T
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
data_train["Attrition"]=label_encoder.fit_transform(data_train["Attrition"])
```

```python
data_train["BusinessTravel"]=label_encoder.fit_transform(data_train["BusinessTravel"])
```

```python
data_train["EducationField"]=label_encoder.fit_transform(data_train["EducationField"])
```

```python
data_train["Gender"]=label_encoder.fit_transform(data_train["Gender"])
```

```python
data_train["Over18"]=label_encoder.fit_transform(data_train["Over18"])
```

```python
data_train["Department"]=label_encoder.fit_transform(data_train["Department"])
```

```python
data_train["JobRole"]=label_encoder.fit_transform(data_train["JobRole"])
data_train["MaritalStatus"]=label_encoder.fit_transform(data_train["MaritalStatus"])
```

```python
data_train.describe().T
```

```python
data_train.isna().sum()
```

```python
data_train.corr()
```

```python
data_train.drop(['BusinessTravel', 'EducationField', 'EmployeeID', 'JobRole', 'EmployeeCount','TotalWorkingYears','NumCompaniesWorked', 'Over18', 'StandardHours'], axis=1, inplace=True)
```

```python
data_train.corr()
```

```python
data_train.isna().sum()
```

```python
data_train.columns
```

```python
y=data_train["Attrition"]
```

```python
x=data_train[['Age', 'Department', 'DistanceFromHome', 'Education',
       'Gender', 'JobLevel', 'MaritalStatus', 'MonthlyIncome',
       'PercentSalaryHike', 'StockOptionLevel', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
```

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y,x1)
Result = Logistic.fit()
```

```python
Result.summary()
```

There is significance difference in monthly income which leads to attrition
