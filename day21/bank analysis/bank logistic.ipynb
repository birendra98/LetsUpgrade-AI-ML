```python
import pandas as pd
```

```python
y=df["Personal Loan"]
```

```python
#income, ccavg,cd account
x=df[['Income','CCAvg','CD Account']]
```

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y,x1)
Result = Logistic.fit()
```

```python
Result.summary()
```

# Inference

There is significance difference between income,ccavg,cd account with respect to approval of personal loan
